## Latihan 1 - Hitung Jarak
**Pseudocode** 
```
    input(v, t)
    S <- v * t
    output(s)
```

**Flowchart**

````python
import schemdraw
from schemdraw.flow import *

with schemdraw.Drawing() as d:
    d+= Start().label("Mulai")
    d+= Arrow().down(d.unit/4)
    
    #Input the string 
    d+= Data(w = 4).label("input:\n (v,t)")
    d+= Arrow().down(d.unit/4)
    
    #Reverse the string
    d+= Box(w = 6).label("s <- v*t")
    d+= Arrow().down(d.unit/4)
    
    #output the string 
    d+= Data(w = 4).label("output:\n (s)")
    d+= Arrow().down(d.unit/4)

    #End program
    d+= (end := Ellipse().label("Selesai"))
    
````

### Code 1A

````python
# Program HitungJarak
# Menghitung jarak berdasarkan masukan kecepatan (v) dan waktu (t)

# KAMUS
# v, t, s : float

# ALGORITMA
v = float(input())
t = float(input())

s = v * t

print(s)
````

### Code 1B

````python
# Program HitungJarak
# Menghitung jarak berdasarkan masukan kecepatan (v) dan waktu (t)

# KAMUS
# v, t, s : float

# ALGORITMA
v = float(input("Masukkan kecepatan = "))
t = float(input("Masukkan waktu = "))

s = v * t

print("Jarak = ", s)
````

## Latihan 2 - Menghitung Luas Segitiga
**Pseudocode** 
```python
    input(alas, tinggi)
    luas <- 0.5 * alas * tinggi
    output(luas)
```

**Flowchart**

````python
import schemdraw
from schemdraw.flow import *

with schemdraw.Drawing() as d:
    d+= Start().label("Mulai")
    d+= Arrow().down(d.unit/4)
    
    #Input the string 
    d+= Data(w = 4).label("input:\n (alas,tinggi)")
    d+= Arrow().down(d.unit/4)
    
    #Reverse the string
    d+= Box(w = 5).label("luas <- 0.5*alas*tinggi")
    d+= Arrow().down(d.unit/4)
    
    #output the string 
    d+= Data(w = 4).label("output:\n (luas)")
    d+= Arrow().down(d.unit/4)

    #End program
    d+= (end := Ellipse().label("Selesai"))
    
````

### Code

````python
# Program MenghitungLuasSegitiga
# Menghitung luas segitiga berdasarkan masukan alas dan tinggi

# KAMUS
# luas, alas, tinggi : float

# ALGORITMA
alas = float(input())
tinggi = float(input())

luas = 0.5 * alas * tinggi

print(luas)
````

## Latihan 3 - MenghitungBiayaBeliKelereng
**Pseudocode** 
```python
    input(m, h, k)
    biaya <- h * 10 + m * 15 + k * 20
    output(biaya)
```

**Flowchart**

````python
import schemdraw
from schemdraw.flow import *

with schemdraw.Drawing() as d:
    d+= Start().label("Mulai")
    d+= Arrow().down(d.unit/4)
    
    #Input the string 
    d+= Data(w = 4).label("input:\n (m,h,k)")
    d+= Arrow().down(d.unit/4)
    
    #Reverse the string
    d+= Box(w = 6.5).label("biaya <- h * 10 + m * 15 + k * 20 ")
    d+= Arrow().down(d.unit/4)
    
    #output the string 
    d+= Data(w = 4).label("output:\n (biaya)")
    d+= Arrow().down(d.unit/4)

    #End program
    d+= (end := Ellipse().label("Selesai"))
    
````

### Code

````python
# Program MenghitungBiayaBeliKelereng;
# Menghitung biaya beli kelereng tergantung jumlah kelereng tiap warna

# KAMUS
# biaya, m, h, k : int

# ALGORITMA
m = int(input()); h = int(input()); k = int(input())

biaya = m * 10 + h * 15 + k * 20

print(biaya)
````
