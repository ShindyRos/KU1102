## Objek python, tipe dasar, dan variabel

Segala sesuatu dalam Python adalah **objek** dan setiap objek dalam Python memiliki **tipe**. Beberapa tipe dasar meliputi:

- **`int`** (integer; bilangan bulat tanpa desimal)
  - `10`
  - `-3`
- **`float`** (float; angka yang memiliki desimal)
  - `7.41`
  - `-0.006`
- **`str`** (string; Urutan karakter yang terlampir dalam kutipan tunggal, kutipan ganda, atau kutipan triple)
  - `'Ini adalah string menggunakan satu kutipan'`
  - `"Ini adalah string menggunakan kutipan ganda"`
  - `'''Ini adalah string yang dikutip triple menggunakan satu kutipan'''`
  - `"""Ini adalah string yang dikutip triple menggunakan kutipan ganda"""`
- **`bool`** (boolean; nilai biner yang _True_ atau _False_)
  - `True`
  - `False`
- **`NoneType`** (tipe khusus yang mewakili tidak adanya nilai)
  - `None`

Di Python, **variabel** adalah nama yang Anda tentukan dalam kode Anda yang memetakan ke **objek** tertentu, **contoh** objek, atau nilai.

Dengan mendefinisikan variabel, kita dapat merujuk pada hal-hal dengan nama yang masuk akal bagi kita. Nama untuk variabel hanya dapat berisi huruf, garis bawah (`_`), atau angka (tidak ada spasi, tanda hubung, atau karakter lain). Nama variabel harus dimulai dengan huruf atau garis bawah.

## Pseudocode dan Flowchart

## Contoh 1 - Roda Pak Pit
**Pseudocode** 
``` python
    input(R)
    Kel <- 2 * 3.14 * R
    output(R)
```
**Flowchart**


````python
import schemdraw
from schemdraw.flow import *

with schemdraw.Drawing() as d:
    d+= Start().label("Mulai")
    d+= Arrow().down(d.unit/4)
    
    #Input the string 
    d+= Data(w = 4).label("input:\n (R)")
    d+= Arrow().down(d.unit/4)
    
    #Reverse the string
    d+= Box(w = 4).label("Kel = 2 * 3.14 * R")
    d+= Arrow().down(d.unit/4)
    
    #output the string 
    d+= Data(w = 4).label("output:\n (Kel)")
    d+= Arrow().down(d.unit/4)

    #End program
    d+= (end := Ellipse().label("Selesai"))
    
````

````python
# Contoh 1 - Roda Pak Pit

# Algoritma
# Program KelilingLingkaran
# Menghitung keliling lingkaran berdasarkan masukan jari-jari

# KAMUS
# R : float
# Kel : float

# ALGORITMA
R = float(input())

Kel = 2 * 3.14 * R

print(Kel)
````

# Contoh 2 - Tinggi Rata-rata
**Pseudocode** 
```python
    input(t1, t2, t3, t4, t5)
    rata <- (t1+t2+t3+t4+t5)/5
    output(rata)
```

**Flowchart**


````python
import schemdraw
from schemdraw.flow import *

with schemdraw.Drawing() as d:
    d+= Start().label("Mulai")
    d+= Arrow().down(d.unit/4)
    
    #Input the string 
    d+= Data(w = 4).label("input:\n (t1, t2, t3, t4, t5)")
    d+= Arrow().down(d.unit/4)
    
    #Reverse the string
    d+= Box(w = 6).label("rata <- (t1+t2+t3+t4+t5)/5")
    d+= Arrow().down(d.unit/4)
    
    #output the string 
    d+= Data(w = 4).label("output:\n (rata)")
    d+= Arrow().down(d.unit/4)

    #End program
    d+= (end := Ellipse().label("Selesai"))
    
````

````python
# Program TinggiRataRata
# Menerima tinggi 5 siswa dan menghitung rata-ratanya

# KAMUS
# t1, t2, t3, t4, t5 : float
# rata : float

# ALGORITMA
t1 = float(input())
t2 = float(input())
t3 = float(input())
t4 = float(input())
t5 = float(input())

rata = (t1 + t2 + t3 + t4 + t5)/5

print (rata)
````
